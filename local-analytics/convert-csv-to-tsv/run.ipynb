# Convert CSV to TSV with chDB

chDB equivalent of the `clickhouse-local` one-liner. Same SQL: `SELECT` from the CSV, write the result `INTO OUTFILE` as `TSVWithNames`. In-process, no server.

Run `./generate.sh` first to create `data/orders.csv`.

```python
import os
import chdb

os.chdir("data")
```

```python
# CSV -> TSV, header kept (TSVWithNames). TRUNCATE makes it idempotent.
chdb.query(
    "SELECT * FROM file('orders.csv') "
    "INTO OUTFILE 'orders_chdb.tsv' TRUNCATE FORMAT TSVWithNames"
)
```

```python
# Read the first rows back to prove it worked.
df = chdb.query("SELECT * FROM file('orders_chdb.tsv', 'TSVWithNames') LIMIT 3", "DataFrame")
print(df.to_string(index=False))
```
