# Convert Parquet to Avro with chDB

In-process ClickHouse, no server. Run `./generate.sh` first to create the sample Parquet files in `./data/`.

Companion article: https://clickhouse.com/resources/engineering/convert-parquet-to-avro

```python
import os
import chdb

os.chdir("data")
```

```python
# Convert Parquet -> Avro: the same SELECT ... FORMAT Avro, into a file.
chdb.query(
    "SELECT * FROM file('events.parquet') "
    "INTO OUTFILE 'events.avro' TRUNCATE FORMAT Avro"
)
print("converted events.parquet -> events.avro")
```

```python
# Read the Avro back and confirm the row count.
chdb.query("SELECT count() AS rows FROM file('events.avro')", "DataFrame")
```

```python
# The nested tuple round-trips as a nested Avro record.
print(
    chdb.query(
        "SELECT event_id, country, device.1 AS device_type, device.2 AS is_even "
        "FROM file('events.avro') ORDER BY event_id LIMIT 3",
        "PrettyCompact",
    )
)
```
