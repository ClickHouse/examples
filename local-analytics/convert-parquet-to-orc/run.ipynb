# Convert Parquet to ORC with chDB

In-process ClickHouse SQL, no server. Run `./generate.sh` first to create `data/events.parquet`.

```python
import os
import chdb

os.chdir("data")
```

Convert Parquet to ORC. The same `SELECT ... INTO OUTFILE ... FORMAT ORC` as the CLI, run from Python.

```python
chdb.query(
    "SELECT * FROM file('events.parquet') "
    "INTO OUTFILE 'events_chdb.orc' TRUNCATE FORMAT ORC"
)
print("wrote events_chdb.orc")
```

Verify: aggregate straight off the ORC file, returned as a pandas DataFrame.

```python
df = chdb.query(
    "SELECT country, count() AS c, round(sum(amount), 2) AS amount "
    "FROM file('events_chdb.orc') GROUP BY country ORDER BY country",
    "DataFrame",
)
df
```
