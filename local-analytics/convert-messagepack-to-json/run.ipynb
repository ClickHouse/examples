# Convert MsgPack to JSON with chDB

Same ClickHouse SQL as the CLI version, run in-process. Run `./generate.sh` first to create `data/events.msgpack`.

```python
import os
import chdb

os.chdir("data")

# MsgPack carries no schema, so the read needs an explicit structure.
STRUCT = "event_id UInt64, ts DateTime, event_type String, country String, amount Float64"
```

```python
sql = (
    f"SELECT * FROM file('events.msgpack', MsgPack, '{STRUCT}') "
    "INTO OUTFILE 'events_chdb.jsonl' TRUNCATE FORMAT JSONEachRow"
)
chdb.query(sql)
```

```python
with open("events_chdb.jsonl") as f:
    for line in list(f)[:3]:
        print(line.rstrip())
```
