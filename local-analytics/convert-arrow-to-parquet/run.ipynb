# Convert Arrow IPC to Parquet with chDB

chDB equivalent of the CLI conversion. Same ClickHouse SQL, in-process, no server.
Run `./generate.sh` first to create `data/events.arrow` and `data/events_stream.arrow`.

```bash
pip install chdb
```

````python
import os
import chdb

os.chdir("data")
````

## Convert Arrow IPC file -> Parquet

The same `SELECT ... INTO OUTFILE ... FORMAT Parquet` you'd run on the CLI.

````python
chdb.query(
    "SELECT * FROM file('events.arrow') "
    "INTO OUTFILE 'events_chdb.parquet' TRUNCATE FORMAT Parquet"
)

print("rows:", chdb.query("SELECT count() FROM file('events_chdb.parquet')", "CSV"), end="")
print(chdb.query("DESCRIBE file('events_chdb.parquet')", "CSV"), end="")
````

## Streaming Arrow IPC needs the `ArrowStream` format

A `.arrow` written by a streaming writer is not an Arrow *file*. Name the framing.

````python
chdb.query(
    "SELECT * FROM file('events_stream.arrow', 'ArrowStream') "
    "INTO OUTFILE 'events_stream_chdb.parquet' TRUNCATE FORMAT Parquet"
)

print("stream rows:", chdb.query("SELECT count() FROM file('events_stream_chdb.parquet')", "CSV"), end="")
````
