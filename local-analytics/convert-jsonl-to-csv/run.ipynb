# Convert JSONL to CSV with chDB

chDB is the in-process ClickHouse engine for Python. Same SQL as `clickhouse local`, no server. Run `./generate.sh` first to create `./data/events.jsonl`.

```python
import os
import chdb

os.chdir("data")
```

Flatten the nested `geo` object and the `tags` array into scalar CSV columns, then write straight to a file with the same `SELECT ... FORMAT`.

```python
sql = """
SELECT
  event_id,
  ts,
  action,
  geo.country AS geo_country,
  geo.city    AS geo_city,
  arrayStringConcat(tags, '|') AS tags
FROM file('events.jsonl')
INTO OUTFILE 'events_chdb.csv' TRUNCATE FORMAT CSVWithNames
"""
chdb.query(sql)

print(open("events_chdb.csv").read())
```
