# Convert CSV to Avro in Python with chDB

Companion to the article [How to convert CSV to Avro](https://clickhouse.com/resources/engineering/convert-csv-to-avro).

Run `./generate.sh` first to create `./data/events.csv`. chDB is the embedded ClickHouse engine for Python: no server, no import step. The Avro schema is derived from the column types chDB infers from the CSV.

```python
import pathlib
import chdb

DATA = pathlib.Path('data')
src = DATA / 'events.csv'
dst = DATA / 'events_chdb.avro'
```

```python
# Convert: read the CSV with file(), write Avro bytes, save to disk.
avro_bytes = chdb.query(
    f"SELECT * FROM file('{src}', 'CSVWithNames') FORMAT Avro"
).bytes()
dst.write_bytes(avro_bytes)
print(f'wrote {dst.name} ({len(avro_bytes)} bytes)')
```

```python
# Read it straight back and confirm the round-trip.
print(chdb.query(f"SELECT * FROM file('{dst}') ORDER BY amount DESC LIMIT 5"))
```

```python
# Inspect the schema carried into the Avro file.
print(chdb.query(f"DESCRIBE file('{dst}')"))
```
