# Convert TSV to CSV with chDB

chDB is ClickHouse as an in-process Python engine. Same SQL as `clickhouse local`, no server. Run `./generate.sh` first to create `data/events.tsv`.

```python
import os
import chdb

os.chdir("data")
```

Convert TSV -> CSV with a header row, writing the file directly from SQL.

```python
chdb.query(
    "SELECT * FROM file('events.tsv') "
    "INTO OUTFILE 'events_chdb.csv' TRUNCATE FORMAT CSVWithNames"
)
print("wrote events_chdb.csv")
```

Read the first few lines back to confirm the header and comma delimiter.

```python
with open("events_chdb.csv") as f:
    for line in list(f)[:6]:
        print(line.rstrip())
```
