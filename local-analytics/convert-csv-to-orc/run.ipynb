# Convert CSV to ORC with chDB

Surface B of [How to convert CSV to ORC](https://clickhouse.com/resources/engineering/convert-csv-to-orc). Run `./generate.sh` first to create `data/events.csv`.

```python
import chdb
```

## Convert CSV -> ORC in one line

Same ClickHouse SQL as the CLI: `SELECT` from the CSV, `INTO OUTFILE` as ORC, in-process.

```python
chdb.query("SELECT * FROM file('data/events.csv') INTO OUTFILE 'data/events_chdb.orc' TRUNCATE FORMAT ORC")
print("wrote data/events_chdb.orc")
```

## Types carried from CSV into ORC

```python
print(chdb.query("DESCRIBE file('data/events_chdb.orc')", "TabSeparated"), end="")
```

## Read it back, or aggregate on the ORC directly

```python
print(chdb.query("SELECT * FROM file('data/events_chdb.orc') LIMIT 5", "CSV"), end="")
print(chdb.query("SELECT count() FROM file('data/events_chdb.orc')", "CSV"), end="")
```
