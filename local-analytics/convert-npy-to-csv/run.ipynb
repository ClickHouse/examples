# Convert NPY to CSV with chDB

Companion to [How to convert NPY to CSV](https://clickhouse.com/resources/engineering/convert-npy-to-csv).

Run `./generate.sh` first to create `./data/signal.npy`. chDB is ClickHouse as an in-process Python engine, so the same SQL runs without a server.

```python
import os
import chdb

os.chdir("data")
```

```python
# 1. Convert a 1D .npy to .csv in one process, no server.
chdb.query(
    "SELECT * FROM file('signal.npy') "
    "INTO OUTFILE 'signal_chdb.csv' TRUNCATE FORMAT CSVWithNames"
)
with open("signal_chdb.csv") as f:
    print(f.read().strip())
```

```python
# 2. Or get the CSV straight back as a string, no file at all.
print(chdb.query("SELECT * FROM file('signal.npy') LIMIT 3 FORMAT CSV", "CSV"))
```
