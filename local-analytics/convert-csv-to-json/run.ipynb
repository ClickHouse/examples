# Convert CSV to JSON with chDB

chDB is the in-process ClickHouse engine for Python. Same SQL as `clickhouse-local`, no server.

Run `./generate.sh` first to create `data/orders.csv`. Requires `pip install chdb`.

```python
import os
import chdb

os.chdir("data")
```

## CSV -> JSON, one object per line (JSONEachRow / NDJSON)

```python
chdb.query(
    "SELECT * FROM file('orders.csv') "
    "INTO OUTFILE 'orders_chdb.jsonl' TRUNCATE FORMAT JSONEachRow"
)

with open('orders_chdb.jsonl') as f:
    for _ in range(3):
        print(f.readline().rstrip())
```

## CSV -> a single JSON array of objects (FORMAT JSON)

```python
print(chdb.query("SELECT * FROM file('orders.csv') LIMIT 2 FORMAT JSON"), end='')
```
