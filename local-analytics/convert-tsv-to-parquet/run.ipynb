# Convert TSV to Parquet with chDB

In-process ClickHouse SQL in Python. No server, no import step. Run `./generate.sh` first to create `data/events.tsv`.

```python
import os
import chdb

os.chdir("data")
```

Convert TSV -> Parquet in one statement. Types are inferred from the TSV and carried into the Parquet schema; ClickHouse compresses with zstd by default.

```python
chdb.query("""
SELECT * FROM file('events.tsv')
INTO OUTFILE 'events_chdb.parquet' TRUNCATE
FORMAT Parquet
""")
```

Confirm the inferred schema landed in the Parquet file.

```python
print(chdb.query("DESCRIBE file('events_chdb.parquet')", "TSV"))
```

Read it back to prove the round-trip is correct.

```python
print(chdb.query("SELECT * FROM file('events_chdb.parquet') ORDER BY event_id LIMIT 5", "CSV"))
```
