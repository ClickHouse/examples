# Convert BSON to CSV with chDB

Read a BSON file, flatten the nested sub-document, write CSV. Same ClickHouse SQL as the CLI, run in-process in Python. Run `./generate.sh` first to create `./data/events.bson`.

```python
import os
import chdb

DATA = os.path.join(os.getcwd(), "data")
src = os.path.join(DATA, "events.bson")
out = os.path.join(DATA, "events_flat_chdb.csv")
```

```python
sql = f"""
SELECT
    event_id,
    event_type,
    geo['city']    AS city,
    geo['country'] AS country,
    amount
FROM file('{src}')
INTO OUTFILE '{out}' TRUNCATE FORMAT CSVWithNames
"""

chdb.query(sql)
print(f"wrote {out}")
print(open(out).read().rstrip())
```
