# Convert NDJSON to Parquet in Python with chDB

Companion to [How to convert NDJSON to Parquet](https://clickhouse.com/resources/engineering/convert-ndjson-to-parquet).

Run `./generate.sh` first to create `./data/events.ndjson`. NDJSON and JSONL are the same format.

```python
import os
import chdb

os.chdir("data")
```

## Convert NDJSON -> Parquet with one SELECT ... INTO OUTFILE

```python
chdb.query(
    "SELECT * FROM file('events.ndjson') "
    "INTO OUTFILE 'events_py.parquet' TRUNCATE FORMAT Parquet"
)
chdb.query("SELECT count() FROM file('events_py.parquet')", "CSV")
```

## Read the nested columns back as a pandas DataFrame

The nested object became a Tuple and the array stayed an Array — both carried into the Parquet file.

```python
chdb.query(
    "SELECT event_id, user.country AS country, user.plan AS plan, items "
    "FROM file('events_py.parquet') ORDER BY event_id LIMIT 5",
    "DataFrame",
)
```
