# Convert JSONL to Parquet with chDB

Companion to [How to convert JSONL to Parquet](https://clickhouse.com/resources/engineering/convert-jsonl-to-parquet).
Run `./generate.sh` first to create `data/events.jsonl`.

```python
import chdb
import os

os.chdir("data")
```

Convert JSONL -> Parquet with the same `SELECT ... INTO OUTFILE ... FORMAT Parquet`, in-process, no server:

```python
chdb.query(
    "SELECT * FROM file('events.jsonl') "
    "INTO OUTFILE 'events_chdb.parquet' TRUNCATE FORMAT Parquet"
)
print("wrote events_chdb.parquet")
```

Verify: same row count, types carried over, nested `device` kept as a Parquet struct:

```python
print(chdb.query("SELECT count() AS rows FROM file('events_chdb.parquet')", "PrettyCompact"))
print(chdb.query("DESCRIBE file('events_chdb.parquet')", "PrettyCompact"))
print(chdb.query(
    "SELECT event_id, device.os AS os, device.ver AS ver, amount "
    "FROM file('events_chdb.parquet') ORDER BY event_id LIMIT 5",
    "PrettyCompact",
))
```
